## Covid Exploratory Data Analysis

```python
import pandas as pd

df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

```python
df = df.iloc[:-1,1:]
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')

plt.barh(total_states,df['positive'],align='center',alpha=0.5,color=(1,0,0), edgecolor=(0.5,0.2,0.5))

plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

```python
figure(num=None,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k')

plt.barh(total_states,df['active'],align='center',alpha=0.5,color=(1,0,0), edgecolor=(0.5,0.2,0.5))

plt.yticks(total_states, df['state_name'])  
plt.xlim(1,max(df['active'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```
